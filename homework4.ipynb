W4A29

```python
s = input()
word = ""
for i in s:
    if 'a' <= i <= 'z':   
        word = word + chr(ord(i) - 32)
    else:                 
        word = word + chr(ord(i) + 32)

print(word)
```

```python
s = input()
hoa = 0
thuong = 0
so = 0
for c in s:
    if 'A' <= c <= 'Z':
        hoa = hoa + 1
    elif 'a' <= c <= 'z':
        thuong = thuong + 1
    elif '0' <= c <= '9':
        so = so + 1
print("Chu hoa:", hoa)
print("Chu thuong:", thuong)
print("Chu so:", so)
```
