```python
c = input()
if c == "A":
    print(" Không có chữ cái liền kề trước a")
else:
    a = c.lower()
    b = chr(ord(a) - 1)
    print(f" Chữ cái liền kề trước {a} là {b}")
```

Bài 6

```python
a = float(input("Cạnh a có giá trị: "))
b = float(input("Cạnh b có giá trị: "))
c = float(input("Cạnh c có giá trị: "))
if a + b > c and a + c > b and b + c > a:
    nuachuvi = (a + b + c) / 2
    dientich = (nuachuvi * (nuachuvi - a) * (nuachuvi - b) * (nuachuvi - c)) ** 0.5
    print("Diện tích tam giác là: ", "%.1f" % dientich)
else:
    print("Không phải là tam giác")
```
