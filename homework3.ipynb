```python
h = int(input())

for i in range(1, h + 1):
    print(" " * (h - i), end="")

    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
```
