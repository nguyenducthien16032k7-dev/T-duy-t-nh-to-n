W6A16 Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n = int(input())
m = int(input())
matrix = []
for i in range(n):
    hang = list(map(int, input().split()))
    matrix.append(hang)
for hang in matrix:    
    for x in hang:
        print(f"{x:>4}", end="")
    print()
```

W6A21 Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện


```python
words = list(map(str, input().split()))
dic = {}
for i, word in enumerate(words):
    if word not in dic:
        dic[word] = []
    dic[word].append(i)
for key in dic:
    dic[key] = tuple(dic[key])  
print(dic)
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 

```python
def zero():
    qua_zero = lst.count(0)
    lst_zero = []
    for i in range(qua_zero):
        lst_zero.append(0)
    for x in lst:
        if x != 0:
            lst_zero.append(x)
    return lst_zero
lst = [0, 2, 9, 0, 0, 0, 9]
zero()
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối


```python
n = list(map(int, input().split()))
sum_cd = 0
for i in range(1, len(n)-1):
    if n[i-1] < n[i] > n[i+1]:
        sum_cd += n[i]
print(sum_cd)
```

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in r

```python
n = list(map(int, input().split()))
x = int(input())
lst = []
for y in n:
    if y < x:
        lst.append(y)
if lst == []:
    print('r')
else:
    print(lst)
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 

```python
n = list(map(int, input().split()))
count = 0
for i in range(len(n)-1):
    if n[i] < n[i+1]:
        count += 1
    elif n[i] > n[i+1]:
        count += 1
    else:
        print("Không là dãy Tăng Giảm")
if len(n)-1 == count and n[0] < n[1]:
    print("Tăng")
elif len(n)-1 == count and n[1] < n[0]:
    print("Giảm")
```
