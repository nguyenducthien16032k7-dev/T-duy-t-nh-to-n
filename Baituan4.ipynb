```python
n = int(input())
step = 0 
while True:
    s = str(n)
    if s == s[::-1]:
        break
    rev = int(s[::-1])
    n = n + rev
    step += 1
print(step, n)
```

```python
n = int(input())
found = False
i = 1
while i * i <= n:
    x = i * i
    s = str(x)
    ok = True
    for j in range(len(s)):
        for k in range(j+1, len(s)):
            if s[j] == s[k]:
                ok = False
    if ok:
        print(x, end=" ")
        found = True
    i += 1
if not found:
    print("no number")
```

W4A10

```python
n = int(input())
max_len = 0
res_x = 1
for i in range(1, n + 1):
    x = i
    count = 0
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3 * x + 1
        count += 1
    if count > max_len:
        max_len = count
        res_x = i
print(res_x, max_len)
```
