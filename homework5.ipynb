Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x

```python
def amstronhg(x):
      if x < 0:
         return False
      else:
         digital = str(x)
         n = len(digital)
         total = 0
         for i in digital:
             total += int(i)**n
      return total == x
x = int(input()) 
print(amstronhg(x))                 
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
def check(n):
    sodoixung = 0
    count = 0
    a = abs(n)
    while True:
        sodoixung += (a % 10) * (10**count)
        count += 1
        if a // 10 == 0:
            break
        a = a // 10
    return sodoixung == abs(n)
n = int(input())
print(check(n))
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except:
        try:
            float(s)
            return 'float'
        except:
            return 'neither'
s = input().strip()
print(detect_number_type(s))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
    
    A = area(quad[0][0], quad[0][1], quad[1][0], quad[1][1], quad[2][0], quad[2][1]) + area(quad[0][0], quad[0][1], quad[2][0], quad[2][1], quad[3][0], quad[3][1])
    A1 = area(p[0], p[1], quad[0][0], quad[0][1], quad[1][0], quad[1][1])
    A2 = area(p[0], p[1], quad[1][0], quad[1][1], quad[2][0], quad[2][1])
    A3 = area(p[0], p[1], quad[2][0], quad[2][1], quad[3][0], quad[3][1])
    A4 = area(p[0], p[1], quad[3][0], quad[3][1], quad[0][0], quad[0][1])
    return A == A1 + A2 + A3 + A4   
quad = [tuple(map(float, input().split())) for _ in range(4)]
p = tuple(map(float, input().split()))  
print(point_in_quadrilateral(quad, p))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero_digits(numbers):
    count = 0
    for n in numbers:
        for digit in str(abs(n)):
            if digit != '0':
                count += 1
    return count

n = list(map(int, input().split()))
print(count_non_zero_digits(n))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def angle_ABC(x1, y1, x2, y2, x3, y3):
    AB = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    BC = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
    AC = math.sqrt((x3 - x1)**2 + (y3 - y1)**2)
    
    if AB == 0 or BC == 0:
        return None 
    
    cos_angle = (AB**2 + BC**2 - AC**2) / (2 * AB * BC)
    angle_rad = math.acos(cos_angle)
    angle_deg = math.degrees(angle_rad)
    
    return round(angle_deg, 2)
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(angle_ABC(x1, y1, x2, y2, x3, y3))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

def binomial_coefficient(n, k):
    if k > n or k < 0:
        return 0
    return factorial(n) // (factorial(k) * factorial(n - k))
n, k = map(int, input().split())
print(binomial_coefficient(n, k))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, 
giá km, phí chờ, phụ phí ban đêm và giảm giá

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 'A':
        base_fare = 5
        km_rate = 2
    elif car_type == 'B':
        base_fare = 3
        km_rate = 1.5
    else:
        return "Invalid car type"
    
    fare = base_fare + km * km_rate + wait_minutes * 0.5
    
    if start_hour >= 22 or start_hour < 6:
        fare *= 1.2
    
    if km > 20:
        fare *= 0.9
    
    return round(fare, 2)
car_type = input("Nhập loại xe A/B: ").strip()
start_hour = int(input("nhập thời gian bắt đầu: "))
km = float(input("nhập khoảng cách (km): "))
wait_minutes = int(input("nhập thời gian chờ (phút): "))
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    total_sum = 0
    for x in range(1, n + 1):
        if len(set(str(x))) == len(str(x)):
            total_sum += x
    return total_sum
n = int(input())
print(sum_1_to_n(n))
```
