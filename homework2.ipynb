Bài 3

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Đây là tam giác đều")
    elif a == b or a == c or b == c:
        print("Đây là tam giác cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Đây là tam giác vuông")    
    else:
        print("Đây là tam giác thường")
    nuachuvi = (a + b + c) / 2
    dientich = (nuachuvi * (nuachuvi - a) * (nuachuvi - b) * (nuachuvi - c)) ** 0.5
    print(f"Diện tích tam giác là {dientich}")


else:
    print("Đây không phải là tam giác")    
```

Bài 5

```python
a, b, c = map(float,input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print("Phương trình vô nghiệm")
elif delta == 0:
    print(f"Phương trình có nghiệm kép x = {-b/(2*a)}")
else:
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print(f"Phương trình có hai nghiệm phân biệt x1 = {x1} và x2 = {x2}")
```
